# Gra w papier kanie≈Ñ nozyce

* **Gra jest odporna na uzytkownik√≥w nie wpisujacych wskazanych znakow**
* **Ca≈Ça rozgrywka jest oparta na losowym generowaniu 3 cyfr, gdzie 0 oznacza papier, 1 oznacza kamie≈Ñ oraz 2 oznacza nozyce**
* **Zasady gdy sa proste zar√≥wno gracz jak i komputer wystawiaja jedna z 3 mozliwo≈õci (papier,kamien,nozyce)**
* **Papier zawsze wygrywa z kamieniem a przegrywa z nozycami**
* **Kamien zawsze wygrywa z nozycami a przegrywa z paierem**
* **Nozyce zawsze wygrywaja z papierem a przegrywaja z kamieniem**

**Powodzenia w grze!**

```python
from time import *
import random

print('Papier, Kamien czy Nozyce?')
sleep(2)
print('Wpisz:')
sleep(1)
print('k - je≈õli kamien')
sleep(1)
print('p - je≈õli papier')
sleep(1)
print('n - jesli nozyce')
sleep(2)
i=True
while i == True:
    ruchK = random.sample(range(3), 1)
    if ruchK[0] == 0:
        K='Papier'
    elif ruchK[0] == 1:
        K='Kamien'
    elif ruchK[0] == 2:
        K='Nozyce'
    a=True
    while a==True:
        print('Uwaga zaczynamy!')
        sleep(1)
        print('3')
        sleep(1)
        print('2')
        sleep(1)
        print('1')
        sleep(1)
        print('START!')
        G = input()  
        if G == 'p':
            G='Papier'
            a=False
        elif G == 'k':
            G='Kamien'
            a=False
        elif G == 'n':
            G='Nozyce'
            a=False
        else:
            print('Wybierz jednƒÖ z moliwo≈õci!')
            sleep(1)
    print(G, 'VS', K )
    sleep(1)
    if (G == 'Kamien' and K == 'Nozyce') or (G == 'Papier' and K == 'Kamien') or (G == 'Nozyce' and K == 'Papier'):
        print('ü•≥ WYGRANA ü•≥')
    elif G == K:
        print('ü§ù REMIS ü§ù')
    else:
        print('üò¢ PRZEGRANA üò¢')
    sleep(1)
    print('Chcesz zagraƒá jeszcze raz ?')
    sleep(1)
    j=True
    while j == True:
        print('Wybierz:')
        sleep(1)
        print('t - jesli TAK')
        sleep(1)
        print('n - jesli NIE')
        b = input()
        if b == 'n':
            i = False
            break
        elif b == 't':
            print('Za chwile nastƒôpna rudna!')
            j= False
            sleep(1.5)
        else:
            print('Podaj t lub n')
            j=True
```
